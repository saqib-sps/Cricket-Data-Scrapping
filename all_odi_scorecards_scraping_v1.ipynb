```python
## function to initialize and configure webdriver
def configure_firefox_driver():
    # add additional Options to the webdriver
    firefox_options = FirefoxOptions()
    # add the argument and make the browser Headless.
    firefox_options.add_argument("--headless")
    # instantiate the Webdriver
    driver = webdriver.Firefox(executable_path=r"geckodriver.exe", options = firefox_options)
    driver.implicitly_wait(3)
    driver.maximize_window()
    
    return driver
```

```python
def get_name_id_from_url(link):
    name_id=link.split("/")
    name_id=name_id[len(name_id)-1].split("-")
    id=name_id[len(name_id)-1]
    name=""
    for i in range(len(name_id)-2):
        name=name+name_id[i]+" "
    name=name+name_id[len(name_id)-2]
    return id,name
```

```python
def get_team_captain (batting_table,start_index =0):
    t1_captain_name = ""
    t1_captain_id = 0
    
    didn_not_bat = None
    # fall_of_wickets = (batting_tables[0].find_element(By.XPATH,"//*[contains (text (), 'Fall of wickets')]"))
    # parent_div = fall_of_wickets. find_element (By.XPATH,"..")
    # print(parent_div.text)
    
    # Total = (batting_tables[0].find_element(By.XPATH,"//*[text () = 'TOTAL']"))
    # parent_div = Total. find_element (By.XPATH,"..")
    # print(parent_div.text)
   #team 1 captain
                   
    rows=(batting_table.find_elements(By.TAG_NAME,"tr"))
    
    
    for i in range(start_index,len(rows)):
        if rows [i].get_attribute ('class') != 'ds-hidden':
            if rows [i].get_attribute ('class')  == "!ds-border-b-0":
                didn_not_bat = rows [i]
            else:
                
                data = rows[i].find_element(By.XPATH,"./td[1]").text
                
                if 'Did not bat' not in data  and 'Fall of wickets' not in data and data != 'Extras' and data != 'TOTAL':
                    link = rows[i].find_element(By.TAG_NAME,"a")
                    if link.get_attribute('href') not in player_profiles_urls:
                        
                        player_profiles_urls.append(link.get_attribute('href'))
                    if '(c)' in data:
                        id, name = get_name_id_from_url(link.get_attribute('href'))
                        
                        t1_captain_name = name
                        captain_id = id
                        
    if t1_captain_name =="":
        spans = didn_not_bat.find_elements(By.TAG_NAME, 'a')
        for s in spans:
            if link.get_attribute('href') not in player_profiles_urls:
                        
                    player_profiles_urls.append(link.get_attribute('href'))
            if '(c)' in s.text:
                
                id, name = get_name_id_from_url(s.get_attribute('href'))
                t1_captain_name = name
                
                captain_id = id
    else:
        if didn_not_bat != None:
            
            spans = didn_not_bat.find_elements(By.TAG_NAME, 'a')
            for s in spans:
                if s.get_attribute('href') not in player_profiles_urls:
                        
                    player_profiles_urls.append(link.get_attribute('href'))

    return t1_captain_name, captain_id
```

```python
def get_match_notes():
    '''
    Purpose: this fucntion is used to get highlights.
    input: it uses the global variable driver.
    output: it returns highlights of a match in the form of dictionary innings_1 and innings 2 are keys and 
    their main points are values. 
    '''
    
    highlights=[]
    innings_1_highlights=[]
    innings_2_highlights=[]
    list1=driver.find_elements(By.XPATH,'//*[@id="main-container"]/div[5]/div[1]/div/div[3]/div[1]/div[2]/div[1]/div[4]/div[2]')
    if len(list1)>0:
        
        lists=list1[0].find_elements(By.TAG_NAME,"ul")
        if (len(lists)>0):
            innings_1_highlights=[]
            list1_items=lists[0].find_elements(By.TAG_NAME,"li")
            for items in list1_items:
                innings_1_highlights.append(items.text)
        if (len(lists)>1):
            innings_2_highlights=[]
            list1_items=lists[1].find_elements(By.TAG_NAME,"li")
            for items in list1_items:
    
                innings_2_highlights.append(items.text)
    else:
        lists=driver.find_elements(By.TAG_NAME,"ul")
        if (len(lists)>0):
            innings_1_highlights=[]
            list1_items=lists[0].find_elements(By.TAG_NAME,"li")
            for items in list1_items:
                innings_1_highlights.append(items.text)
        if (len(lists)>1):
            innings_2_highlights=[]
            list1_items=lists[1].find_elements(By.TAG_NAME,"li")
            for items in list1_items:
    
                innings_2_highlights.append(items.text)
        
    
    highlights.append(innings_1_highlights)
    highlights.append(innings_2_highlights)
    return highlights
```

```python
def bubbleSort(arr):
    n = len(arr)
    swapped = False
    # Traverse through all array elements
    for i in range(n-1):
        for j in range(0, n-i-1):
            if arr[j] > arr[j + 1]:
                swapped = True
                arr[j], arr[j + 1] = arr[j + 1], arr[j]         
        if not swapped:
            return
```
