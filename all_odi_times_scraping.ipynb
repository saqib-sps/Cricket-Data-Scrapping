```python
# importing required libraries
import pandas as pd
from selenium import webdriver
from selenium.webdriver.firefox.options import Options as FirefoxOptions
from selenium.webdriver.common.by import By
import sys
from selenium.webdriver import Firefox
from selenium.webdriver.firefox.options import Options
from selenium import webdriver
#import webdriver_manager
#from webdriver_manager.firefox import GeckoDriverManager
```

```python
## function to initialize and configure webdriver
def configure_firefox_driver():
    # add additional Options to the webdriver
    firefox_options = FirefoxOptions()
    # add the argument and make the browser Headless.
    firefox_options.add_argument("--headless")
    # instantiate the Webdriver
    driver = webdriver.Firefox(executable_path=r"geckodriver.exe", options = firefox_options)
    driver.implicitly_wait(3)
    driver.maximize_window()
    
    return driver
```

```python
def get_match_details():
    '''
    Purpose: If there is a match that has incomplete game then this fucntion is used to get match related details.
    input: it uses the global variable driver.
    output: it returns details of a match in the form of dictionary in which match number acts as key and other details 
    of the match are value.
    '''
    match_details={}
   
    
    
   
    tables=driver.find_elements(By.TAG_NAME,"table")
    if len(tables)>0:
        for i in range(len(tables)):
            if tables[i].get_attribute("class")=="ds-w-full ds-table ds-table-sm ds-table-auto ":
                rows=tables[i].find_elements(By.TAG_NAME,"tr")
                for j in range(len(rows)):
                    
                    tds=rows[j].find_elements(By.TAG_NAME,"td")
                    if (len(tds)>0):
                        
                        if tds[0].text=="Toss":
                            table=tables[i]
                            rows=table.find_elements(By.TAG_NAME,"tr")
                            match_details['ground']=rows[0].text
                            ground_id=(rows[0].find_element(By.TAG_NAME,'a')).get_attribute('href')
                            for k in range(1,len(rows)):
                    
                                tds=rows[k].find_elements(By.TAG_NAME,"td")
                                for data in range(0,len(tds),2):                                       
                                    match_details[tds[data].text]=tds[data+1].text
                                    
                            
                           
                            return match_details
```

```python

details_match_ids=[]
match_days=[]
start_times = []


def set_match_details (complete_details):

    if 'Match number' in complete_details.keys():
        n=complete_details['Match number']
        n=n.split(" ")
        details_match_ids.append(n[len(n)-1])
    else:
        details_match_ids.append('None')
        
    if 'Match days' in complete_details.keys():
        match_days.append(complete_details['Match days'])
    else:
        match_days.append('-')
    if 'Hours of play (local time)' in complete_details.keys():
        start_times.append(complete_details['Hours of play (local time)'])
    else:
        start_times.append('-')
```

```python
def scrap_all_scorecards(url):
    '''
    Purpose: this fucntion is used to scrap scorecard of each match using other helping functions.
    input: it takes address of a scorecard and then accesses it using the driver to scrap other information.
    output: this funciton returns scorecards of a match in the form of dictionary where match number is key and other
    information of the scorecard is value.
    '''
                   
    driver.get(url)
    complete_details = get_match_details()
    set_match_details(complete_details)
    
    
```

```python
def create_dataframes():
     
    details=pd.DataFrame({'match_id':details_match_ids,'match_days':match_days,'match_start_time':start_times

                          })
    
    details.to_csv('odi_match_time_details.csv') 
                
    
    
    
```

```python
def bubbleSort(arr):
    n = len(arr)
    swapped = False
    # Traverse through all array elements
    for i in range(n-1):
        for j in range(0, n-i-1):
            if arr[j] > arr[j + 1]:
                swapped = True
                arr[j], arr[j + 1] = arr[j + 1], arr[j]         
        if not swapped:
            return
```

```python
#ODI_Batting
match_ids = odi_bat_df["match_ids"].value_counts().index.tolist()
bubbleSort(match_ids)

at_home = []

for i in range(0, len(match_ids)): 
    df = odi_bat_df[odi_bat_df['match_ids']== match_ids[i]]
    df
    arr = []
    arr.append(df["team"].iloc[0])
    arr.append(df["oppostion"].iloc[0])
    team1_home = 0
    team2_home = 0

    if (arr[0] in df["home_team"].iloc[0]):
        team1_home = 1
    elif (arr[1] in df["home_team"].iloc[0]):
        team2_home = 1
    else:
        team1_home = 2
        team2_home = 2

    for k in range(0, len(df)):
        if df["team"].iloc[k] == arr[0]:
            at_home.append(team1_home)
        elif df["team"].iloc[k] == arr[1]:
            at_home.append(team2_home)
            
odi_bat_df["at_home"] = at_home
```

```python
#ODI_Bowling
match_ids = odi_ball_df["match_id"].value_counts().index.tolist()
bubbleSort(match_ids)

at_home = []

for i in range(0, len(match_ids)): 
    df = odi_ball_df[odi_ball_df['match_id']== match_ids[i]]
    df
    arr = []
    arr.append(df["bowler_team"].iloc[0])
    arr.append(df["bowler_opposition"].iloc[0])
    team1_home = 0
    team2_home = 0

    if (arr[0] in df["home_team"].iloc[0]):
        team1_home = 1
    elif (arr[1] in df["home_team"].iloc[0]):
        team2_home = 1
    else:
        team1_home = 2
        team2_home = 2

    for k in range(0, len(df)):
        if df["bowler_team"].iloc[k] == arr[0]:
            at_home.append(team1_home)
        elif df["bowler_team"].iloc[k] == arr[1]:
            at_home.append(team2_home)
            
odi_ball_df["at_home"] = at_home
```
